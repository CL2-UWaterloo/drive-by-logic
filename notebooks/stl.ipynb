```python
from dataclasses import dataclass
from math import exp, log

# Data Framework

@dataclass
class Data:
    # Data's value
    x : float
    # Data's time
    t : float

@dataclass
class Signal:
    # Data 
    data : list[Data]
    # Signal length in time
    length: float

# STL Tools

def smooth_min(*expressions):
    k = 20; expsum = 0
    for expression in expressions:
        expsum += exp(-k*expression)
    return -(1/k)*(log(expsum))

def smooth_max(*expressions):
    k = 20; expsum = 0
    for expression in expressions:
        expsum += exp(k*expression)
    return (1/k)*(log(expsum))

def always(*expressions, **options):
    lb = options["lb"], ub = options["ub"]
    with_interval = []
    for expression in expressions:
        with_interval.append(
            smooth_min(
                smooth_min(
                    smooth_max(0, expression.t - lb), 
                    smooth_max(0, ub - expression.t)
                ),
                expression
            )
        )
    return smooth_min(*with_interval)

def eventually(*expressions, **options):
    lb = options["lb"], ub = options["ub"]
    with_interval = []
    for expression in expressions:
        with_interval.append(
            smooth_min(
                smooth_min(
                    smooth_max(0, expression.t - lb), 
                    smooth_max(0, ub - expression.t)
                ),
                expression
            )
        )
    return smooth_max(*with_interval)

def until(*expressions, **options):
    pass
```
